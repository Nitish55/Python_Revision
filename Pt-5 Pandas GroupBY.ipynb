# Merging

```python
df1 = pd.DataFrame({'key1': ['K0', 'K8', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                    'C': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
df2 = pd.DataFrame({'key2': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})
print(df1)
print("-"*55)
print(df2)
```

```python
# Inner Join on different key name, Also we can rename key columna s common name
mg=pd.merge(df1,df2,how='inner',left_on='key1',right_on='key2')
mg
```

```python
# Rename On commomn key
df1["key2"]=df1["key1"]
df1.drop("key1",axis=1,inplace=True)
df1
```

```python
nmg=pd.merge(df1,df2,how='inner',on='key2')
nmg
```

```python
nmg=pd.merge(df1,df2,how='left',on='key2')
nmg
```

```python
print(df1)
print("-"*55)
print(df2)
```

```python
nmg=pd.merge(df1,df2,how='outer',on='key2')
nmg
```

# On multiple Keys

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
print(left)
print("-"*55)
print(right)
```

```python
mr=pd.merge(left,right,how='inner',on=['key1','key2'])
mr
```

```python
# On multiple merge
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2','K1'],
                     'key2': ['K0', 'K1', 'K0', 'K1','K0'],
                        'A': ['A0', 'A1', 'A2', 'A3','A4'],
                        'B': ['B0', 'B1', 'B2', 'B3','B4']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2','K1'],
                               'key2': ['K0', 'K0', 'K0', 'K0','K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3','C5'],
                                  'D': ['D0', 'D1', 'D2', 'D3','Df5']})
print(left)
print("-"*55)
print(right)
```

```python
mr=pd.merge(left,right,how='inner',on=['key1','key2'])
mr
```
