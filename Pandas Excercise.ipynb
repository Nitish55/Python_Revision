# Which continent drinks more beer on average?

```python
df=pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/drinks.csv",sep=",")
df
```

```python
gr=df.groupby(["beer_servings","continent"])
gr=gr.sum().sort_values("beer_servings",ascending=False)
gr.head(1)
```

# Since we have to find the Avergae

```python
gr=df.groupby(["continent"])
gr=gr["beer_servings"].mean()
gr
```
