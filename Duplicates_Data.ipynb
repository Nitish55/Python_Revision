```python
import pandas as pd
  
# List of Tuples
employees = [('Stuti', 28, 'Varanasi'),
             ("Stuti",28,"Varanasi"),
             ("Stuti",28,"Delhi"),
            ('Saumya', 32, 'Delhi'),
            ('Aaditya', 25, 'Mumbai'),
            ('Saumya', 32, 'Delhi'),
            ('Saumya', 32, 'Delhi'),
            ('Saumya', 32, 'Mumbai'),
            ('Aaditya', 40, 'Dehradun'),
            ('Seema', 32, 'Delhi')
            ]
  
# Creating a DataFrame object
df = pd.DataFrame(employees, 
                  columns = ['Name', 'Age', 'City'])
  
# Print the Dataframe
df
```

### Checing Duplicates

```python
dup=df[df.duplicated()]
dup
```

## USing Keep

```python
dup=df[df.duplicated(keep="first")]
dup
```

```python
dup=df[df.duplicated(keep="last")]
dup
```

```python
dup=df[df.duplicated(keep=False)]
dup
```

## Using selected columns

```python
dup=df[df.duplicated(subset="Name",keep="last")]
dup
```

# Dropping duplicates

```python
df
```

```python
df1=df.drop_duplicates(keep="first")
df1
```

```python
df1.drop_duplicates(subset="Name",keep="first")
```
